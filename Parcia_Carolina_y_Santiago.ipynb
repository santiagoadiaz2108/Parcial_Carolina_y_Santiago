Punto 1

```python
import pandas as pd

data = {
    'Producto': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Ventas': [120, 340, 290, 410, 150, 380],
    'Costo': [80, 200, 180, 260, 90, 240],
    'Categoria': ['Tecnología', 'Hogar', 'Tecnología', 'Hogar', 'Oficina', 'Tecnología'],
    'Mes': ['Enero', 'Enero', 'Febrero', 'Febrero', 'Marzo', 'Marzo']
}

df = pd.DataFrame(data)

import numpy as np

ventas_np = np.array(df['Ventas'])

promedio = ventas_np.mean()
maximo = ventas_np.max()
minimo = ventas_np.min()

promedio, maximo, minimo
```

mean(): calcula el promedio de las ventas.

max(): identifica el valor más alto dentro del arreglo.

min(): identifica el valor más bajo dentro del arreglo.

Punto 2

```python
ventas_mayor_300 = np.where(ventas_np > 300, 1, 0)
ventas_mayor_300
```

Punto 3

```python
df['Ganancia'] = df['Ventas'] - df['Costo']

df.head()

```

```python
df[df['Ganancia'] > 100][['Producto', 'Ganancia']]
```

Punto 4

```python
filtro_tecnologia = df[
    (df['Categoria'] == 'Tecnología') & (df['Ventas'] > 300)
][['Producto', 'Ventas', 'Ganancia']]

filtro_tecnologia
```

Punto 5

```python
resumen_mes = df.groupby('Mes').agg(
    Ventas_totales=('Ventas', 'sum'),
    Ganancia_promedio=('Ganancia', 'mean')
)

resumen_mes
```

sum(): suma todos los valores de un grupo.

mean(): calcula el promedio de un grupo.

Punto 6

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(df['Producto'], df['Ventas'])
plt.title('Ventas por Producto')
plt.xlabel('Producto')
plt.ylabel('Ventas')
plt.xticks(rotation=0)
plt.show()
```

Punto 7

```python
plt.figure()
plt.plot(df['Producto'], df['Ventas'], label='Ventas')
plt.plot(df['Producto'], df['Costo'], label='Costo')
plt.title('Comparación de Ventas y Costos por Producto')
plt.xlabel('Producto')
plt.ylabel('Valor')
plt.legend()
plt.show()
```

Tendencia: Las ventas son consistentemente mayores que los costos, lo que indica rentabilidad en todos los productos, aunque con diferentes márgenes.

Punto 8

```python
import seaborn as sns

plt.figure()
sns.boxplot(x='Categoria', y='Ventas', data=df)
plt.title('Distribución de Ventas por Categoría')
plt.show()
```

La categoría Tecnología presenta mayor variabilidad, ya que tiene un rango más amplio de valores de ventas, evidenciado por un boxplot más extensos.

Punto 9

```python
plt.figure()
sns.scatterplot(x='Ventas', y='Ganancia', hue='Categoria', data=df)
plt.title('Relación entre Ventas y Ganancia')
plt.show()
```

Se observa una relación positiva: a mayores ventas, mayor ganancia, aunque la pendiente varía según la categoría, lo que indica diferencias en estructura de costos.
